```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D
from matplotlib.legend_handler import HandlerTuple
import matplotlib.ticker as ticker


plt.rcParams["figure.figsize"] = [4.5,3]
plt.rcParams["figure.dpi"] = 300
plt.rc('font', size=10)
```

```python
snr_gw40 = pd.read_csv('data/snr_40_gw.csv')
t_gw40 = snr_gw40['t_merge']
snrgw40 = snr_gw40['snr']

snr_df40_2_1200 = pd.read_csv('data/snr_40_df_2_1200.csv')
t_df40_2_1200 = snr_df40_2_1200['t_merge']
snrdf40_2_1200 = snr_df40_2_1200['snr']

snr_df40_25_1200 = pd.read_csv('data/snr_40_df_2.5_1200.csv')
t_df40_25_1200 = snr_df40_25_1200['t_merge']
snrdf40_25_1200 = snr_df40_25_1200['snr']

snr_df40_3_1200 = pd.read_csv('data/snr_40_df_3_1200.csv')
t_df40_3_1200 = snr_df40_3_1200['t_merge']
snrdf40_3_1200 = snr_df40_3_1200['snr']

snr_df40_35_1200 = pd.read_csv('data/snr_40_df_3.5_1200.csv')
t_df40_35_1200 = snr_df40_35_1200['t_merge']
snrdf40_35_1200 = snr_df40_35_1200['snr']
```

```python
plt.plot(t_gw40, snrgw40, color='k',linewidth = 1,label="plunge orbit")

plt.plot(t_df40_2_1200, snrdf40_2_1200, color='#00FF00',linewidth = 1,linestyle='dashed', dashes =(4,5), label="plunge orbit")
plt.plot(t_df40_25_1200, snrdf40_25_1200, color='#0066FF',linewidth = 1,linestyle='dashed',label="plunge orbit")
plt.plot(t_df40_3_1200, snrdf40_3_1200, color='#FFA500',linewidth = 1,linestyle='dashed',label="plunge orbit")
plt.plot(t_df40_35_1200, snrdf40_35_1200, color='#FF0000',linewidth = 1,linestyle='dashed' ,label="plunge orbit")

plt.xscale('log')
plt.yscale('log')
plt.xlim(3e5,1e1)
plt.ylim(10,1e6)

color_handles = [
    Line2D([0], [0], color='#00FF00', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#0066FF', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#FFA500', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#FF0000', lw=0.8, linestyle='dashed'),
]

color_legend = plt.legend(
    handles=color_handles,
    labels=[r"$\gamma = 2.0$", r"$\gamma = 2.5$", r"$\gamma = 3.0$", r"$\gamma = 3.5$"],
    loc="upper left",                  # 
    bbox_to_anchor=(0.01, 0.99),        # x=2%y=2%
    frameon=False,
    borderaxespad=0.5,                  # 
    borderpad=0.5,
    fontsize=9                       # 
)

plt.text(
    x=0.3, y=0.9,  
    s=r'm = $40M_{\odot}$',  
    transform=plt.gca().transAxes,  
    fontsize=9,
)

plt.tight_layout()


plt.xlabel(r'Time to merge(yr)', fontdict={'family' : 'Times New Roman'})
plt.ylabel(r'SNR', fontdict={'family' : 'Times New Roman'})

ax = plt.gca()
ax.yaxis.set_minor_locator(ticker.LogLocator(base=10.0, subs=np.arange(2, 10) * 0.1, numticks=10))
ax.xaxis.set_minor_locator(ticker.LogLocator(base=10.0, subs=np.arange(2, 10) * 0.1, numticks=10))

plt.savefig('D:/pyfile/25DMEMRIs/output/snr_40_1200.pdf', bbox_inches='tight')
plt.show()
```
