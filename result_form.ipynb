```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D
from matplotlib.legend_handler import HandlerTuple
import matplotlib.ticker as ticker

plt.rcParams["figure.figsize"] = [4.5,3]
plt.rcParams["figure.dpi"] = 300
plt.rc('font', size=10)
```

```python
formgw = pd.read_csv('data/40emri_form_gw0.26.csv')
a_gw = formgw['a_decay_dominate']
oneminuse_gw = formgw['1-e']
a0 = formgw['a_losscone']

formdf40_2_1200=pd.read_csv('data/40_2_1200_emri_form_df0.26.csv')
a_df40_2_1200 = formdf40_2_1200['a_decay_dominate']
oneminuse_df40_2_1200 = formdf40_2_1200['1-e']

formdf40_25_1200=pd.read_csv('data/40_2.5_1200_emri_form_df0.26.csv')
a_df40_25_1200 = formdf40_25_1200['a_decay_dominate']
oneminuse_df40_25_1200 = formdf40_25_1200['1-e']

formdf40_3_1200=pd.read_csv('data/40_3_1200_emri_form_df0.26.csv')
a_df40_3_1200 = formdf40_3_1200['a_decay_dominate']
oneminuse_df40_3_1200 = formdf40_3_1200['1-e']

formdf40_35_1200=pd.read_csv('data/40_3.5_1200_emri_form_df0.26.csv')
a_df40_35_1200 = formdf40_35_1200['a_decay_dominate']
oneminuse_df40_35_1200 = formdf40_35_1200['1-e']
```

```python
formgw10 = pd.read_csv('data/10emri_form_gw0.26.csv')
a_gw10 = formgw10['a_decay_dominate']
oneminuse_gw10 = formgw10['1-e']
a010 = formgw10['a_losscone']

formdf10_2_1200 = pd.read_csv('data/10_2_1200_emri_form_df0.26.csv')
a_df10_2_1200 = formdf10_2_1200['a_decay_dominate']
oneminuse_df10_2_1200 = formdf10_2_1200['1-e']

formdf10_25_1200 = pd.read_csv('data/10_2.5_1200_emri_form_df0.26.csv')
a_df10_25_1200 = formdf10_25_1200['a_decay_dominate']
oneminuse_df10_25_1200 = formdf10_25_1200['1-e']

formdf10_3_1200 = pd.read_csv('data/10_3_1200_emri_form_df0.26.csv')
a_df10_3_1200 = formdf10_3_1200['a_decay_dominate']
oneminuse_df10_3_1200 = formdf10_3_1200['1-e']

formdf10_35_1200 = pd.read_csv('data/10_3.5_1200_emri_form_df0.26.csv')
a_df10_35_1200 = formdf10_35_1200['a_decay_dominate']
oneminuse_df10_35_1200 = formdf10_35_1200['1-e']
```

```python

plt.plot(oneminuse_gw, a0, color='#668B8B',label="plunge orbit")
plt.plot(oneminuse_gw, a_gw, linewidth = 1, color='k',label="critical orbit without dm")



plt.plot(oneminuse_df40_2_1200, a_df40_2_1200, linewidth =1, color='#00FF00',linestyle='dashed' )
plt.plot(oneminuse_df40_25_1200, a_df40_25_1200, linewidth = 1, color='#0066FF',linestyle='dashed' )
plt.plot(oneminuse_df40_3_1200, a_df40_3_1200, linewidth = 1, color='#FFA500',linestyle='dashed' )
plt.plot(oneminuse_df40_35_1200, a_df40_35_1200, linewidth = 1, color='#FF0000',linestyle='dashed' )

plt.fill_between(oneminuse_gw, a0, color='#668B8B', alpha = 0.5)

plt.xscale('log')
plt.yscale('log')
plt.xlim(3e-7,1e-1)
plt.ylim(1e-6,7e-1)
plt.xlabel(r'$\rm{1-e}$', fontdict={'family' : 'Times New Roman'})
plt.ylabel(r'$\rm{a(pc)}$', fontdict={'family' : 'Times New Roman'})

color_handles = [
    Line2D([0], [0], color='#00FF00', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#0066FF', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#FFA500', lw=0.8, linestyle='dashed'),
    Line2D([0], [0], color='#FF0000', lw=0.8, linestyle='dashed'),
]

color_legend = plt.legend(
    handles=color_handles,
    labels=[r"$\gamma = 2.0$", r"$\gamma = 2.5$", r"$\gamma = 3.0$", r"$\gamma = 3.5$"],
    loc="lower left",                  # 基准点设为左下角
    bbox_to_anchor=(0.01, 0.01),        # 调整位置：x=2%左边界，y=2%下边界
    frameon=False,
    borderaxespad=0.5,                  # 图例与轴边界的间距
    borderpad=0.5,
    fontsize=9                       # 图例边框内边距
)

plt.gca().add_artist(color_legend)

plt.text(
    x=0.3, y=0.08,  
    s=r'm = $40M_{\odot}$',  
    transform=plt.gca().transAxes,  
    fontsize=9,
)


plt.tight_layout()

plt.xticks([10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1)])
plt.yticks([10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1),10**(0)])

ax = plt.gca()
ax.yaxis.set_minor_locator(ticker.LogLocator(base=10.0, subs=np.arange(2, 10) * 0.1, numticks=10))
ax.xaxis.set_minor_locator(ticker.LogLocator(base=10.0, subs=np.arange(2, 10) * 0.1, numticks=10))

plt.savefig('D:/pyfile/25DMEMRIs/output/emri_form_40_1200.pdf', bbox_inches='tight')
plt.show()
```
